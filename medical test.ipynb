```python
df_numeric=df.apply(pd.to_numeric,errors='coerce')
```

```python
#this is heat map pearson correlation matrix
corrmat = df_numeric.corr(numeric_only=True)
f, ax = plt.subplots(figsize=(16, 12))
sns.heatmap(corrmat, vmax=0.8, square=True);
plt.title('Pearson Correlation Matrix Heatmap',fontsize=16)
plt.show()
df_coy = df.copy()
```

* The heat map displays a weak negative correlation between age and blood sugar levels.
    
* Overall, the majority of the data points on the heatmap indicate blood sugar rates below 0.

* The distribution suggests that as age increases, there tends to be a slight decrease in blood sugar levels, although the correlation is not strong.

* There may be some variability in blood sugar levels across different age groups, but the general trend indicates lower blood sugar levels across the board.

```python
plt.figure(figsize=(20,20))
plt.title('Pearson Correlation of Features', size = 15)
colormap = sns.diverging_palette(10, 220, as_cmap = True)
sns.heatmap(df_numeric.corr(),
            cmap = colormap,
            square = True,
            annot = True,
            linewidths=0.1,vmax=1.0, linecolor='white',
            annot_kws={'fontsize':12 })
plt.show()
```
